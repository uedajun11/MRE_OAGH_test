```python
import scipy.io as sio
import torch
import numpy as np
import os


root = os.path.dirname(os.getcwd())

src = root+incl_input+measures[3]
dst = src+"/pt_files/"

os.makedirs(dst, exist_ok=True)

files_only = [
    f for f in os.listdir(src)
    if f.endswith(".mat") and os.path.isfile(os.path.join(src, f))
]
print(len(files_only))
for f in files_only:
    mat = sio.loadmat(os.path.join(src, f))

    wave = torch.from_numpy(mat["wave"]).float()        # [256,256,8]
    wave = wave.unsqueeze(0)                            # [1,256,256,8]

    mu = torch.from_numpy(mat["mu"].astype(np.float32))            # [256,256]

    mfre = torch.from_numpy(np.array(mat["mfre"]).squeeze()).double()
    index = torch.from_numpy(mat["index"].astype(np.int64))       # [1,1]
    
    sample = {
        "wave": wave,
        "mu": mu,
        "mfre": mfre,
        "index": index
    }

    torch.save(sample, os.path.join(dst, f.replace(".mat", ".pt")))
```
